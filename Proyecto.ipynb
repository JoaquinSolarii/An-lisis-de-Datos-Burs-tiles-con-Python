```python
import yfinance as yf
import pandas as pd
import json
import matplotlib.pyplot as plt
```

## Using the yfinance Library to Extract Stock Data


Using the `Ticker` module we can create an object that will allow us to access functions to extract data. To do this we need to provide the ticker symbol for the stock, here the company is Apple and the ticker symbol is `AAPL`.


```python
apple = yf.Ticker("AAPL")
```

```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json
```

### Stock Info


```python
with open('apple.json') as json_file:
    apple_info = json.load(json_file)
apple_info
```

```python
apple_info['country']
```

### Extracting Share Price


```python
apple_share_price_data = apple.history(period="max")
```

```python
apple_share_price_data.head()
```

```python
apple_share_price_data.reset_index(inplace=True)
```

```python
apple_share_price_data.plot(x="Date", y="Open")
```

### Extracting Dividends


```python
apple.dividends
```

```python
apple.dividends.plot()
```

## Exercise — AMD (Advanced Micro Devices)


Now using the `Ticker` module create an object for AMD (Advanced Micro Devices) with the ticker symbol `AMD`; name the object `amd`.


```python
# Creamos el objeto Ticker para AMD
amd = yf.Ticker("AMD")
```

```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/amd.json
```

```python
with open('amd.json') as json_file:
    amd_info = json.load(json_file)
amd_info
```

**Question 1** — Use the key `'country'` to find the country the stock belongs to.


```python
# Pregunta 1: País al que pertenece la acción de AMD
amd_info['country']
```

**Question 2** — Use the key `'sector'` to find the sector the stock belongs to.


```python
# Pregunta 2: Sector al que pertenece AMD
amd_info['sector']
```

**Question 3** — Obtain stock data for AMD using the `history` function, set the `period` to max. Find the `Volume` traded on the first day (first row).


```python
# Pregunta 3: Descargamos todo el historial de AMD
amd_data = amd.history(period="max")
amd_data.reset_index(inplace=True)
amd_data.head()
```

```python
# Volumen negociado en el primer día disponible
volumen_primer_dia = amd_data['Volume'].iloc[0]
print(f"Volumen en el primer día de cotización de AMD: {volumen_primer_dia:,}")
```

---
## Visualizaciones adicionales — AMD


```python
# Gráfico del precio de apertura a lo largo del tiempo
amd_data.plot(x="Date", y="Open", title="AMD — Precio de Apertura Histórico", figsize=(12, 5), color="steelblue")
plt.xlabel("Fecha")
plt.ylabel("Precio de Apertura (USD)")
plt.tight_layout()
plt.show()
```

```python
# Gráfico comparativo Open vs Close
fig, ax = plt.subplots(figsize=(12, 5))
amd_data.plot(x="Date", y="Open", ax=ax, label="Apertura (Open)", color="steelblue")
amd_data.plot(x="Date", y="Close", ax=ax, label="Cierre (Close)", color="tomato", linestyle="--")
ax.set_title("AMD — Precio de Apertura vs Cierre")
ax.set_xlabel("Fecha")
ax.set_ylabel("Precio (USD)")
plt.tight_layout()
plt.show()
```

```python
# Gráfico del volumen negociado
amd_data.plot(x="Date", y="Volume", title="AMD — Volumen Negociado Histórico",
              figsize=(12, 4), color="mediumpurple", alpha=0.7, kind="area")
plt.xlabel("Fecha")
plt.ylabel("Volumen")
plt.tight_layout()
plt.show()
```

```python
# Dividendos de AMD
amd.dividends.plot(title="AMD — Dividendos Históricos", figsize=(12, 4), color="seagreen")
plt.ylabel("Dividendo por acción (USD)")
plt.tight_layout()
plt.show()
```
